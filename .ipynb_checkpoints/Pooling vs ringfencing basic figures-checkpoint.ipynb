```python
##The purpose of this notebook is to generate the graphs of pooling vs ringfencing for the uniform and normal case.

#Imports
from tqdm import tqdm_notebook as tqdm
tqdm().pandas()
%matplotlib inline
import tkinter
import pacal as pa
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
from skimage import measure
import matplotlib.colors as colors
```

```python
#Uniform Distribution Pooling Vs Ringfencing
x = np.linspace(-5, 5, 1000)
z = pa.UniformDistr(5 - 10, 5 + 10)
z_2 = 0.5*(z + z)
z.plot(label = 'Ring-fenced', color = 'black')
z_2.plot(color='green', label ='Pooled')
#z_2.plot(color='green', label ='Pooled')
plt.legend(loc = 'best', fontsize = 13)
plt.title('Uniformly Distributed Payoffs, $\gamma = 5, \delta = 20$', fontsize= 20)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.savefig('UniformDisPoolvsRingfence.pgf')
```

```python
#Normal pooling vs ring-fencing
y = pa.NormalDistr(5, 3)
y_2 = 0.5*(y + y)
y.plot(label = 'Ring-fenced')
y_2.plot(color='green', label ='Pooled')
plt.legend(loc = 'best', fontsize = 13)
plt.title('Normally Distributed Payoffs, $\mu = 5, \sigma = 3$', fontsize= 20)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.savefig('NormalDisPoolvsRingfence.pgf')
```
