```python
#The Purpose of this Notebook is to generate the payoffs vs correlation contour plots for the uniform and normal cases.

#Imports and Parameter Setting
from tqdm import tqdm_notebook as tqdm
tqdm().pandas()
%matplotlib inline
import pacal as pa
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from skimage import measure
import matplotlib.colors as colors

mean = 5
```

```python

# set the colormap and centre the colorbar
class MidpointNormalize(colors.Normalize):
	"""
	Normalise the colorbar so that diverging bars work there way either side from a prescribed midpoint value)

	e.g. im=ax1.imshow(array, norm=MidpointNormalize(midpoint=0.,vmin=-100, vmax=100))
	"""
	def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
		self.midpoint = midpoint
		colors.Normalize.__init__(self, vmin, vmax, clip)

	def __call__(self, value, clip=None):
		# I'm ignoring masked values and all kinds of edge cases to make a
		# simple example...
		x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
		return np.ma.masked_array(np.interp(value, x, y), np.isnan(value))

#Determines payoff from ringfencing two issues
def E_Ringfencing(Dist1, Dist2, plot = False, ub = 20, lb = -20, spacing = 1e-4):
    Dist1Pos = Dist1 | pa.Gt(0)
    Dist2Pos = Dist2 | pa.Gt(0)
    
    EV_Ring = (Dist1.ccdf_value(0)) * (Dist1.ccdf_value(0) * (Dist1Pos.mean())) + (Dist2.ccdf_value(0)) * (Dist2.ccdf_value(0) * (Dist2Pos.mean()))

    if plot:
        grid = np.arange(lb, ub, spacing)
        plt.plot(grid, Dist1.pdf(grid), label = 'Distribution 1')
        plt.plot(grid, Dist2.pdf(grid), label = 'Distribution 2')
        plt.legend(loc='best'), plt.suptitle('PDFs')

    return EV_Ring
```

```python
#Plotting axes for contour plot.
Z = np.load('UniformCorrelation.npy')
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='PiYG', edgecolor='none')
ax.set_xlabel('mean')
ax.set_ylabel('Variance')
ax.set_zlabel('Expected Value')



ax.set_title('surface');
plt.show()
```

```python
#Creating Contour plot
contours = measure.find_contours(Z, 0)

plt.figure(figsize = (10, 10))


plt.imshow(Z, extent=[X.min(), X.max(), Y.min(), Y.max()], origin='lower', cmap = 'RdBu', norm=MidpointNormalize(midpoint=0))
#plt.colorbar(im, cax=cax, orientation='vertical')

cb = plt.colorbar()
cb.ax.tick_params(labelsize= 18)

# for n, contour in enumerate(contours):
#     ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

plt.title('Expected Pooling Gains for Uniform Distribution, $\gamma = 5$ ', fontsize = 20)

plt.xlabel('$\delta$ (Support Size)', fontsize = 20)

plt.ylabel('$ \\theta$ (Correlation Parameter)', fontsize = 20)

plt.xticks(fontsize =  20)
plt.yticks(fontsize = 20)



plt.savefig('UniformCorrelation.pgf')
```
