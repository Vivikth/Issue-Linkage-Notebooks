```python
#The purpose of this notebook is to create contour payoff plots in the identical normal and uniform cases

#imports
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.colors as colors
import numpy as np
import pacal as pa
from mpl_toolkits.mplot3d import Axes3D  
from skimage import measure
%matplotlib inline
```

```python
# set the colormap and centre the colorbar
class MidpointNormalize(colors.Normalize):
	"""
	Normalise the colorbar so that diverging bars work there way either side from a prescribed midpoint value)

	e.g. im=ax1.imshow(array, norm=MidpointNormalize(midpoint=0.,vmin=-100, vmax=100))
	"""
	def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
		self.midpoint = midpoint
		colors.Normalize.__init__(self, vmin, vmax, clip)

	def __call__(self, value, clip=None):
		# I'm ignoring masked values and all kinds of edge cases to make a
		# simple example...
		x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
		return np.ma.masked_array(np.interp(value, x, y), np.isnan(value))

def Expected_Pooling_Gain(distributions):
    '''Function that computes expected gain from pooling for number of distributions
    '''
    Poslist = []
    EV_Ring = 0
    for indx, dist in enumerate(distributions):
        Poslist.append(dist | pa.Gt(0))
        if indx == 0:
            Conv_Dist = dist
        else:
            Conv_Dist += dist
        EV_Ring += (dist.ccdf_value(0)) * (dist.ccdf_value(0) * (Poslist[indx].mean()))
    Conv_DistPos = Conv_Dist | pa.Gt(0)
    V_Pool = Conv_Dist.ccdf_value(0) * Conv_DistPos.mean() * Conv_Dist.ccdf_value(0)

    return V_Pool - EV_Ring

#A faster version of the above function, specifically coded for 2-issue uniform and 2-issue normal distributions.
def E_ID_Gain_Pooling(mean, std = 1, δ = 1, distribution = 'Uniform' ):
    if distribution != 'Normal' and distribution != 'Uniform':
        raise ValueError("Invalid Distribution")
        
    elif distribution == 'Normal':
#         if not mean - δ < 0:
#             return 0
        Dist1 = pa.NormalDistr(mean, std)
        Dist2 = pa.NormalDistr(mean, std)
    elif distribution == 'Uniform':
        if not mean - δ < 0:
            return 0
        Dist1 = pa.UniformDistr(mean - δ, mean + δ)
        Dist2 = pa.UniformDistr(mean - δ, mean + δ)
        
    else:
        raise ValueError('Should not be possible to end up here')
     
    Dist1Pos = Dist1 | pa.Gt(0)
    Dist2Pos = Dist2 | pa.Gt(0)

    
    EV_Ring = (Dist2.ccdf_value(0)) * (Dist1.ccdf_value(0) * (Dist1Pos.mean())) + (Dist2.ccdf_value(0)) * (Dist1.ccdf_value(0) * (Dist2Pos.mean()))
    
    if distribution == 'Normal':
        Conv_Dist =  pa.NormalDistr(2 * mean, (2 * (std**2))**0.5)
    else:
        Conv_Dist = Dist1 + Dist2
    Conv_DistPos = Conv_Dist | pa.Gt(0)
    V_Pool = Conv_Dist.ccdf_value(0) * Conv_DistPos.mean() * Conv_Dist.ccdf_value(0)

    return V_Pool - EV_Ring
```
